```python
import numpy as np

print(np.triu(np.ones(df_assembly.shape)).astype(np.bool)) #visualiza o triângulo superior

df = df_assembly.where(np.triu(np.ones(df_assembly.shape)).astype(np.bool))
df

```

```python
df = df.stack().reset_index()
df.columns = ['Species_1','Species_2','Identity']
df
```

```python
df_no_self_comparison = df[(df["Species_1"] != df["Species_2"])] #remove comparações entre mesma sequência
```

```python
#função para extrair número de acesso
def extract_accession(name):
    accession_list = name.split("_")[0:2]
    accession = "_".join(accession_list)
    return(accession)

teste = "NZ_LR134338.1_rrna_17"
extract_accession(teste)
```

```python
df_no_self_comparison["accession_species_1"] = df_no_self_comparison["Species_1"].apply(extract_accession)
df_no_self_comparison["accession_species_2"] = df_no_self_comparison["Species_2"].apply(extract_accession)
df_no_self_comparison
```

```python
#avalia se as colunas accession_species_1 e accession_species_2 possuem valores iguais (intra) ou diferentes (inter)
df_no_self_comparison['Category'] = np.where(df_no_self_comparison['accession_species_1'] == df_no_self_comparison["accession_species_2"], 1, 0) #1 = intra, 0 = inter
```

```python
#Primeira possibilidade de verificar o melhor threshold
import matplotlib.pyplot as plt
import numpy as np

score = np.array(df_no_self_comparison["Identity"])
y = np.array(df_no_self_comparison["Category"])

# false positive rate
fpr = []
# true positive rate
tpr = []
# Iterate thresholds from 0.0, 0.01, ... 1.0
thresholds = np.arange(0.0, 1.001, .001)

# get number of positive and negative examples in the dataset
P = sum(y)
N = len(y) - P

# iterate through all thresholds and determine fraction of true positives
# and false positives found at this threshold
for thresh in thresholds:
    FP=0
    TP=0
    for i in range(len(score)):
        if (score[i] >= thresh):
            if y[i] == 1:
                TP = TP + 1
            if y[i] == 0:
                FP = FP + 1
    fpr.append(FP/float(N))
    tpr.append(TP/float(P))

plt.scatter(fpr, tpr)
plt.show()
```

```python
listas = [thresholds, tpr, fpr]
df_roc = pd.DataFrame(listas)
```

```python
df_roc = df_roc.T
```

```python
df_roc.columns = ["Threshold", "TPR", "FPR"]
```

```python
df_roc
```

```python
optimal_idx = np.argmax(df_roc["TPR"] - df_roc["FPR"])
optimal_threshold = thresholds[optimal_idx]
```

```python
#Segunda opção

import numpy as np
from sklearn import metrics


y_scores = np.array(df_no_self_comparison["Identity"])
y_true = np.array(df_no_self_comparison["Category"])

fpr, tpr, thresholds = metrics.roc_curve(y_true, y_scores)

```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_true, y_scores)
```

```python
import pylab as pl

####################################
# The optimal cut off would be where tpr is high and fpr is low
# tpr - (1-fpr) is zero or near to zero is the optimal cut off point
####################################
i = np.arange(len(tpr)) # index for df
roc = pd.DataFrame({'fpr' : pd.Series(fpr, index=i),'tpr' : pd.Series(tpr, index = i), '1-fpr' : pd.Series(1-fpr, index = i), 'tf' : pd.Series(tpr - (1-fpr), index = i), 'thresholds' : pd.Series(thresholds, index = i)})
roc.ix[(roc.tf-0).abs().argsort()[:1]]

# Plot tpr vs 1-fpr
fig, ax = pl.subplots()
pl.plot(roc['tpr'])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.title('Receiver operating characteristic')
ax.set_xticklabels([])
```
